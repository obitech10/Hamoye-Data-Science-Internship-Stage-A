# QUESTION 9

#### COMPARE PERCENTAGE CHANGE IN FUEL COST PER UNIT BURNED IN 1998 TO THAT OF 1994

```python
fuel_data_1988 = fuel_data[fuel_data['fuel_type_code_pudl']== "coal"]
fuel_data_1998 = fuel_data[fuel_data['report_year']==1998]
x= fuel_data_1998['fuel_cost_per_unit_burned'].mean()
print("Average of Fuel Cost per Unit Burned of Coal in 1998=", round(x,0))
print(" ")

fuel_data_1994 = fuel_data[fuel_data['fuel_type_code_pudl']== "coal"]
fuel_data_1994 = fuel_data[fuel_data['report_year']==1994]
y = fuel_data_1994['fuel_cost_per_unit_burned'].mean()
print("Average of Fuel Cost per Unit Burned of Coal in 1994=", round(y,0))
print(" ")

z=((x-y)/y)*(100)
print("Percentage change in Fuel Cost per Unit Burned of Coal in 1998 to that of 1994 =", round(z,0),"%")
```
